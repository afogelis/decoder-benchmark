# Decoder Comparison: MWPM vs Union-Find vs Belief Propagation

We benchmark three surface-code decoders on the same syndrome batches and compare them on
accuracy, runtime and memory. The headline finding -- belief propagation is dominated by
matching-based decoders on surface codes -- reproduces the decoder literature.

```python
%matplotlib inline
import matplotlib.pyplot as plt

from decbench import (
    BenchmarkConfig,
    build_leaderboard,
    format_leaderboard,
    run_benchmark,
)
from decbench.viz import plot_accuracy_vs_p, plot_pareto

config = BenchmarkConfig(
    decoders=["mwpm", "union_find", "bp"],
    distances=[3, 5],
    error_rates=[0.005, 0.008, 0.01, 0.012],
    shots=4000,
    seed=2026,
)
result = run_benchmark(config)
print(format_leaderboard(build_leaderboard(result)))
```

## Accuracy vs runtime (Pareto frontier)

Decoders on the lower-left frontier are both fast and accurate. MWPM (optimised C++) and the
union-find decoder define the accuracy frontier; BP sits off it.

```python
plot_pareto(result)
plt.tight_layout()
plt.show()

plot_accuracy_vs_p(result, distance=5)
plt.tight_layout()
plt.show()
```
